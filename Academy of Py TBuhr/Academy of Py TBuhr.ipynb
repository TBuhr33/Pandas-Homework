```python
#Load in CSV files and set up dependencies
import pandas as pd
import numpy as np

schools_csv = "PyCitySchools_Schools.csv"
students_csv = "PyCitySchools_Scores.csv"
```

```python
#Read into the CSV's and store as pandas dataframes (df)
school_df = pd.read_csv(schools_csv)
student_df = pd.read_csv(students_csv)
```

```python
# merge the two CSV's into a single data frame 
complete_df = pd.merge(school_df, student_df, how = "left", on = ["school_name", "school_name"])
```

# District Summary

```python
#Create a df for the district to perform analysis
Dist_df = {}
Dist_df = pd.DataFrame()

#create headers for df table, pull data from csv
Dist_df["Total Schools"] = [school_df.school_name.value_counts().sum()]
Dist_df["Total Students"] = [school_df["size"].sum()]
Dist_df["Total Budget"] = [school_df["budget"].sum()]
Dist_df["Average Math Score"] = [student_df["math_score"].mean()]
Dist_df["Average Reading Score"] = [student_df["reading_score"].mean()]

#determine passing rate equations, passing is >= 70
Total_Students = len(student_df["Student ID"])
math_passing = len(student_df[student_df["math_score"] >= 70])/ Total_Students
reading_passing = len(student_df[student_df["reading_score"] >= 70])/ Total_Students

overall_pass = (math_passing + reading_passing)/2

#insert the passing rates into the df
Dist_df["%Passing Math"] = math_passing
Dist_df["%Passing Reading"] = reading_passing
Dist_df["%Overall Passing Rate"] = overall_pass

#redeclare the df with the new data
Dist_df = pd.DataFrame(Dist_df)

#display the df/table
Dist_df

```

# School Summary

```python
#Group schools, also by type
Schools = complete_df.set_index('school_name').groupby(['school_name'])
types = school_df.set_index('school_name')['type']

#find students per school
stud_per_school = complete_df.groupby('school_name')['Student ID'].count()

#find total budget per school 
budg_per_school = school_df.set_index('school_name')['budget']

#find budget per student at each school
budg_perst_persch = school_df.set_index('school_name')['budget']/school_df.set_index('school_name')['size']

#average math test score by school
mean_math_score = Schools['math_score'].mean()

#average reading test score by school 
mean_read_score = Schools['reading_score'].mean()

#passing percentage of math by school (pass is >=70)
pass_math = complete_df[complete_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stud_per_school 

#passing percantage of reading by school (>=70)
pass_read = complete_df[complete_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stud_per_school

#overall/average passing grade by school across
overall = complete_df[(complete_df['reading_score'] >= 70) & (complete_df['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/stud_per_school

#create df/dict to correspond table headers with resources above
schSumm = pd.DataFrame({"School Type": types,
                       "Total Students": stud_per_school,
                       "Budget Per School": budg_per_school,
                       "Budget Per Student": budg_perst_persch,
                       "Average Math Score": mean_math_score,
                       "Average Reading Score": mean_read_score,
                       "% Passing Math": pass_math,
                       "% Passing Reading": pass_read,
                       "% Overall Passing": overall})

#set up the table keys to output
schSumm = schSumm[['School Type', 'Total Students', 'Budget Per School', 'Budget Per Student', 'Average Math Score',
                  'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]

schSumm
```

# Top Performing Schools

# Scores by School Spending (per student)

```python
#create the spending bins
bins = [0, 585, 615, 645, 675]
bin_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#create a data set for the spending bins
complete_df['spending_bins'] = pd.cut(complete_df['budget']/complete_df['size'], bins, labels = bin_names)

#group the df by spending 
group_spending = complete_df.groupby('spending_bins')

#find the average math and reading scores from the new category/df
math_avg = group_spending['math_score'].mean()
reading_avg = group_spending['reading_score'].mean()

#find the passing percentage for the new category/df
math_passing = complete_df[complete_df['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/group_spending['Student ID'].count()
reading_passing = complete_df[complete_df['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/group_spending['Student ID'].count()
overall_passing = (math_passing + reading_passing)/2

#set up the new df for scores by school spending
spend_scores = pd.DataFrame({"Average Math Score": math_avg, "Average Reading Score": reading_avg, "% Passing Math": math_passing,
    "% Passing Reading": reading_passing,"% Overall Passing": overall_passing})
 
#set up table keys for output
spend_scores = spend_scores[['Average Math Score', 'Average Reading Score', '%Passing Math', '% Passing Reading', "% Overall Passing"]]

#rename the spending bins index 
spend_scores.index.name = "Spending Per Student"

spend_scores
```

# Scores by School Size

```python
#repeat similar steps as the school spending
bins = [0, 1000, 2000, 5000]
bin_names = ["Small (<1000)", "Medium (1000-2000)" , "Large (>2000)"]
complete_df['size_bins'] = pd.cut(complete_df['size'], bins, labels = bin_names)

#group by size
size_score = complete_df.groupby('size_bins')

#find for size instead of spending
math_avg_size = size_score['math_score'].mean()
reading_avg_size = size_score['reading_score'].mean()

math_passing_size = complete_df[complete_df['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/size_score['Student ID'].count()
reading_passing_size = complete_df[complete_df['reading_score'] >= 70].groupby('size_bins')['Student ID'].count()/size_score['Student ID'].count()
overall_passing_size = (math_passing_size + reading_passing_size)/2

size_score = pd.DataFrame({"Average Math Score": math_avg_size, "Average Reading Score": reading_avg_size, "% Passing Math": math_passing_size,
    "% Passing Reading": reading_passing_size,"% Overall Passing": overall_passing_size})

size_score = size_score[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', "% Overall Passing"]]

size_score.index.name = "School Size"

size_score
```
